```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Tu código para leer los datos aquí...

# Columnas de interés para el ajuste lineal
x_columns = ['V_BE', 'V_BE1', 'V_BE2']
y_columns = ['I_B', 'I_B1', 'I_B2']

plt.figure(figsize=(8, 6))  # Tamaño de la figura (opcional)

# Iterar a través de cada par de columnas
for i in range(len(x_columns)):
    x = data[x_columns[i]]
    y = data[y_columns[i]]

    # Filtrar datos por encima de x=0.65
    x_filtered = x[x > 0.65]
    y_filtered = y[x > 0.65]

    # Realizar el ajuste lineal solo para los datos filtrados
    coefficients = np.polyfit(x_filtered, y_filtered, 1)
    poly = np.poly1d(coefficients)

    # Calcular la intersección con el eje x (donde y = 0)
    x_intercept = np.roots(poly - np.array(0))
    x_intercept_real = x_intercept[np.isreal(x_intercept)].real

    # Crear una secuencia de puntos para la línea de ajuste
    x_line = np.linspace(min(x), max(x), 100)
    y_line = poly(x_line)

    plt.grid()

    # Graficar todos los puntos
    plt.scatter(x, y, label=f'Datos {i+1}')

    # Graficar solo los puntos filtrados
    plt.scatter(x_filtered, y_filtered)

    # Graficar la recta de ajuste solo para los datos filtrados
    plt.plot(x_line, y_line, label=f'Ajuste {i+1}')

    # Marcar la intersección con el eje x
    plt.scatter(x_intercept_real, np.zeros_like(x_intercept_real), label=f'Intersección {i+1}', marker='x', s=100)

# Etiquetas y título
plt.xlabel('Etiqueta eje X')
plt.ylabel('Etiqueta eje Y')
plt.title('Gráfico de Dispersión con Ajuste Lineal para Datos Filtrados')

plt.ylim(-0.00001, max(y))  # Establece el límite inferior en -0.00001 y el límite superior en el máximo de los datos en y

# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
