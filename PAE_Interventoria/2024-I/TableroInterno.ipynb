```python
df_OR = pd.read_csv('Datos/Reporte análisis organoléptico.csv', sep=';', encoding='latin-1')
df_OR.replace('null', np.nan, inplace=True)

df_OR = df_OR[~df_OR['Códigovisita'].str.startswith('PR')]

df_OR
```

```python
#Vamos a trabajar cuales instituciones incumplen ciertos item
#Primero filtramos solo las IE's
print(df_Resultados.columns)

df_filtrado_IE = df_Resultados[df_Resultados['Código'].str.startswith('IE')]

df_filtrado_OR = df_Resultados[df_Resultados['Código'].str.startswith('OR')]

df_filtrado_BO = df_Resultados[df_Resultados['Instrumento'].str.startswith('Bodega')]

df_filtrado_VH = df_Resultados[df_Resultados['Instrumento'].str.startswith('Vehículo')]


df_filtrado_OR
```

```python
import pandas as pd

# Crear un diccionario de mapeo: Item -> Pregunta
item_to_pregunta = dict(zip(df_filtrado_IE['Item'], df_filtrado_IE['Pregunta']))

# Crear un nuevo dataframe para almacenar los resultados
result_df = pd.DataFrame()

# Iterar a través de las columnas de df_IE
for col in df_IE.columns:
    try:
        # Extraer el número de la columna (asumiendo el formato "1 P", "2 P", ...)
        num = int(col.split(' ')[0])
        
        # Multiplicar por 10 para obtener el Item correspondiente
        item = num * 10
        
        # Buscar la pregunta correspondiente usando el diccionario
        pregunta = item_to_pregunta.get(item, None)
        
        # Agregar la información al nuevo dataframe
        result_df[col] = [pregunta]
    except ValueError:
        # Si ocurre un ValueError, se ignora la columna
        print(f"Se ignoró la columna: {col}")

# Transponer el dataframe para que las columnas sean los enunciados
result_df = result_df.T
#result_df.columns = ['Pregunta']

# Guardar el DataFrame resultante como un archivo Excel
result_df.to_excel('resultado.xlsx', index=True)

result_df


```

## traspocision de tablas relacionales (No ejecutar porque se vuelven a voltear)


```python
import pandas as pd

df_relacional_BO = pd.read_excel('RelacionPregunta-Nu-Item_BO.xlsx')

# se traspone 
relacional_BO_traspuesta = df_relacional_BO.transpose()
#print(relacional_BO_traspuesta)

#Se fija la primera columna como indice
relacional_BO_traspuesta.columns = relacional_BO_traspuesta.iloc[0] 

#Luego se elimina la primera fila
relacional_BO_traspuesta = relacional_BO_traspuesta[1:]

#Se establece la primera fila como indice
relacional_BO_traspuesta = relacional_BO_traspuesta.set_index(relacional_BO_traspuesta.columns[0])

#guardar como tipo excel
output_path = 'RelacionPregunta-Nu-Item_BO.xlsx'

relacional_BO_traspuesta.to_excel(output_path)

```

```python
#se repite el proceso para los vehiculos
df_relacional_VH = pd.read_excel('RelacionPregunta-Nu-Item_VH.xlsx')

#se traspone
relacional_VH_traspuesta = df_relacional_VH.transpose()

#se fija la primera columna como encabezado
relacional_VH_traspuesta.columns = relacional_VH_traspuesta.iloc[0]

#se elimina la primer fila
relacional_VH_traspuesta = relacional_VH_traspuesta[1:]

#se pone como indice la primera fila
relacional_VH_traspuesta = relacional_VH_traspuesta.set_index(relacional_VH_traspuesta.columns[0])

#Se guarda como tipo excel
output_path = 'RelacionPregunta-Nu-Item_VH.xlsx'
relacional_VH_traspuesta.to_excel(output_path)
```
