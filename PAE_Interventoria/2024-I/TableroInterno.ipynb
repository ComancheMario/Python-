```python
df_Resultados = pd.read_csv('Datos/Resultados .csv', sep= ';', encoding='latin-1')
df_Resultados.replace('null', np.nan, inplace=True)
#df_Resultados.dropna()

df_Resultados
```

```python
#Vamos a trabajar cuales instituciones incumplen ciertos item
#Primero filtramos solo las IE's
print(df_Resultados.columns)

df_filtrado_IE = df_Resultados[df_Resultados['Código'].str.startswith('IE')]

df_filtrado_BO = df_Resultados[df_Resultados['Instrumento'].str.startswith('Bodega')]

df_filtrado_VH = df_Resultados[df_Resultados['Instrumento'].str.startswith('Vehículo')]

df_filtrado_BO
```

```python
import pandas as pd

# Crear un diccionario de mapeo: Item -> Pregunta
item_to_pregunta = dict(zip(df_filtrado_IE['Item'], df_filtrado_IE['Pregunta']))

# Crear un nuevo dataframe para almacenar los resultados
result_df = pd.DataFrame()

# Iterar a través de las columnas de df_IE
for col in df_IE.columns:
    try:
        # Extraer el número de la columna (asumiendo el formato "1 P", "2 P", ...)
        num = int(col.split(' ')[0])
        
        # Multiplicar por 10 para obtener el Item correspondiente
        item = num * 10
        
        # Buscar la pregunta correspondiente usando el diccionario
        pregunta = item_to_pregunta.get(item, None)
        
        # Agregar la información al nuevo dataframe
        result_df[col] = [pregunta]
    except ValueError:
        # Si ocurre un ValueError, se ignora la columna
        print(f"Se ignoró la columna: {col}")

# Transponer el dataframe para que las columnas sean los enunciados
result_df = result_df.T
#result_df.columns = ['Pregunta']

# Guardar el DataFrame resultante como un archivo Excel
result_df.to_excel('PAE_Interventoria/2024-I/.xlsx', index=True)

result_df


```
