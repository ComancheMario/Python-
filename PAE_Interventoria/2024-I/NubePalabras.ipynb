```python
#Stop word en español
stop_words = set(stopwords.words('spanish'))
#Extraer sustantivos
def extraer_sustantivos(texto):
    tokens = nltk.word_tokenize(texto.lower())
    tags = nltk.pos_tag(tokens)
    sustantivos = [word for word, pos in tags if pos in ['NN','NNS','NNP','NNPS']  and word not in stop_words]
    return ' '.join(sustantivos)

df_filtrado['Sustantivos'] = df_filtrado['Observación'].apply(extraer_sustantivos)

df_filtrado
```
