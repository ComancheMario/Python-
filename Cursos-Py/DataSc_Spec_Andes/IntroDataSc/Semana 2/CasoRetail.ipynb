```python
#Elimino datos nulos
data=data.dropna(subset=['review_scores_rating'])

#data['review_scores_rating'].astype(int)

#creo los intervalos
data['Interval']= pd.cut(data['review_scores_rating'], bins=list(range(20,data['review_scores_rating'].astype(int).max()+6, 6)),right=False)


#tabla de frecuencia
frec_data = data['Interval'].value_counts().sort_index()

#Agrego la columna de frecuencia relativa
frec_data['frec_relativa']=frec_data/frec_data.sum()

#Agrego la columna frecuencia relativa en porcentaje
frec_data['frec_relativa_porcentaje']=frec_data['frec_relativa'].apply(lambda x: x*100)

#Agrego la columna frecuencia acumulada
frec_data['frec_acumulada']=frec_data['frec_relativa'].cumsum()

#Agrego la columna frecuencia acumulada en porcentaje
frec_data['frec_acumulada_porcentaje']=frec_data['frec_acumulada'].apply(lambda x: x*100)

frec_data
```
