## Graficos de barras


```python
import pandas as pd
import plotly.express as px

#cargar datos


data=pd.read_csv('Ordenes_productos_C1_M2.3.csv', sep=';', encoding='latin-1')
data.head(5)
```

```python
#tabla de frecuencias para el grafico
data_estado = data['departamento_vendedor'].value_counts().reset_index()

#se cambia el nombre de las columnas y se hace el grafico de barras
data_estado.columns = ['departamento_vendedor','conteo']

fig = px.bar(data_estado, x='departamento_vendedor', y='conteo')
fig.show()
```

## Grafico barras agrupadas

```python
#Arreglar datos
data_producto = data.groupby(['departamento_vendedor','nombre_categoria_producto']).count().reset_index()[['orden_id','departamento_vendedor','nombre_categoria_producto']]

#Cambiar nombre columnas
data_producto = data_producto.rename(columns={'orden_id':'conteo'})

#filtro por producto
data_producto = data_producto[data_producto['nombre_categoria_producto']=='Celulares']

#grafico
fig = px.bar(data_producto, x='nombre_categoria_producto', y='conteo', color='departamento_vendedor', barmode='group')

fig.show()


```

## Grafico barras apiladas

```python

#Arreglar datos 
data_estado = data.groupby(['departamento_vendedor','nombre_categoria_producto']).count()[['orden_id']].reset_index()

#cambiar nombre  de columnas
data_estado = data_estado.rename(columns={'orden_id':'conteo'})

#grafico
fig = px.bar(data_estado,x='departamento_vendedor',y='conteo',color='nombre_categoria_producto')
fig.show()
```

## Grafico barras apiladas

```python
#arreglar los datos
data_normalizada = data.groupby(['departamento_vendedor','nombre_categoria_producto']).agg({'orden_id':'count'})
state = data.groupby(['departamento_vendedor']).agg({'orden_id':'count'})

#Normalizar datos
data_normalizada = data_normalizada.div(state, level='departamento_vendedor').reset_index()

#cambiar nombre de columnas
data_normalizada = data_normalizada.rename(columns={'orden_id':'conteo'})

fig = px.bar(data_normalizada, x='departamento_vendedor',y='conteo',color='nombre_categoria_producto')
fig.show()
```

```python
data=pd.read_csv('listings.csv')

#buscar 'review_scores_rating'

if 'review_scores_rating' in data.columns:
    review_scores_rating = data['review_scores_rating']
else:
    raise KeyError('No se encuentra la variable solicitada')
```

```python
#Elimino datos nulos
data=data.dropna(subset=['review_scores_rating'])

#data['review_scores_rating'].astype(int)

#creo los intervalos
data['Interval']= pd.cut(data['review_scores_rating'], bins=list(range(20,data['review_scores_rating'].astype(int).max()+6, 6)),right=False)


#tabla de frecuencia
frec_data = data['Interval'].value_counts().sort_index()

#Agrego la columna de frecuencia relativa
frec_data['frec_relativa']=frec_data/frec_data.sum()

#Agrego la columna frecuencia relativa en porcentaje
frec_data['frec_relativa_porcentaje']=frec_data['frec_relativa'].apply(lambda x: x*100)

#Agrego la columna frecuencia acumulada
frec_data['frec_acumulada']=frec_data['frec_relativa'].cumsum()

#Agrego la columna frecuencia acumulada en porcentaje
frec_data['frec_acumulada_porcentaje']=frec_data['frec_acumulada'].apply(lambda x: x*100)

frec_data
```
