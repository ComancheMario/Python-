```python
import numpy as np
import matplotlib.pyplot as plt

# Define la función de susceptibilidad en función de la frecuencia
def susceptibilidad(frecuencia):
    # Aquí puedes poner tu expresión analítica para la susceptibilidad
    # Por ejemplo, vamos a usar una susceptibilidad de la forma chi(f) = A / (f^2 - f_0^2 + 1j * gamma * f)
    A = 1.0
    f0 = 10.0
    gamma = 0.1
    return A / (frecuencia**2 - f0**2 + 1j * gamma * frecuencia)

# Crea un rango de frecuencias
frecuencias = np.linspace(0.1, 20, 1000)

# Calcula la susceptibilidad para cada frecuencia en el rango
susceptibilidades = susceptibilidad(frecuencias)

# Grafica la parte real e imaginaria de la susceptibilidad
plt.plot(frecuencias, np.real(susceptibilidades), label='Parte Real')
plt.plot(frecuencias, np.imag(susceptibilidades), label='Parte Imaginaria')
plt.xlabel('Frecuencia')
plt.ylabel('Susceptibilidad')
plt.title('Grafico de la Susceptibilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define la función de susceptibilidad en función de la frecuencia angular (omega)
def susceptibilidad(omega, omega_p, omega_0, Q):
    suscept = (omega_p)**2 / (((omega_0)**2 - (omega)**2) + 2j * (omega_0 / Q) * omega)
    return suscept.real + 1j * np.abs(suscept.imag)

# Parámetros
omega_p = 1.0  # Frecuencia de plasma
omega_0 = 5.0  # Frecuencia de resonancia
Q = 10.0       # Factor de calidad

# Rango de frecuencias angulares
omega_values = np.linspace(0, 10, 1000)

# Calcula la susceptibilidad para cada frecuencia angular en el rango
susceptibility_values = susceptibilidad(omega_values, omega_p, omega_0, Q)

# Grafica la parte real e imaginaria de la susceptibilidad
plt.plot(omega_values, np.real(susceptibility_values), label='Parte Real')
plt.plot(omega_values, np.imag(susceptibility_values), label='Parte Imaginaria')
plt.xlabel('Frecuencia Angular (omega)')
plt.ylabel('Susceptibilidad')
plt.title('Grafico de la Susceptibilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
M = 1.0  # Valor de M
omega = 1.0  # Frecuencia angular
Omega_r_values = np.linspace(0, 2, 1000)  # Valores de Omega_r

# Rango de tiempo
t = np.linspace(0, 2, 1000)

# Función para calcular el valor esperado del operador dipolar
def dipole_operator_expected_value(Omega_r, t):
    return -M * np.sin(2 * Omega_r * t) * np.sin(omega * t)

# Calcula los valores esperados para cada valor de Omega_r
dipole_values = [dipole_operator_expected_value(Omega_r, t) for Omega_r in Omega_r_values]

# Grafica los valores esperados en función de Omega_r*t
plt.plot(Omega_r_values * t[-1], dipole_values)
plt.xlabel('Omega_r * t')
plt.ylabel('Valor esperado del operador dipolar')
plt.title('Valor esperado del operador dipolar respecto a Omega_r * t')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función para el valor esperado del operador dipolar
def dipolar_value(rabi_freq, time):
    # Por ejemplo, aquí podrías tener una función ficticia que devuelve un valor basado en la frecuencia de Rabi y el tiempo
    return np.sin(rabi_freq * time) * np.exp(-time)
    

# Definir el rango de valores para la frecuencia de Rabi y el tiempo
rabi_freq_range = np.linspace(0, 10, 100)  # Cambia los límites y la cantidad de puntos según sea necesario
time_range = np.linspace(0, 100, 100)  # Cambia los límites y la cantidad de puntos según sea necesario

# Calcular el valor esperado del operador dipolar para cada combinación de frecuencia de Rabi y tiempo
dipolar_values = np.zeros((len(rabi_freq_range), len(time_range)))
for i, rabi_freq in enumerate(rabi_freq_range):
    for j, time in enumerate(time_range):
        dipolar_values[i, j] = dipolar_value(rabi_freq, time)

# Graficar
plt.figure(figsize=(10, 6))
plt.contourf(rabi_freq_range, time_range, dipolar_values, cmap='viridis')
plt.colorbar(label='Valor Esperado del Operador Dipolar')
plt.xlabel('Frecuencia de Rabi')
plt.ylabel('Tiempo')
plt.title('Valor Esperado del Operador Dipolar en Oscilaciones de Rabi')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función para el valor esperado del operador dipolar
def dipolar_value(M, Omega_r, omega_eg, time):
    return -M * np.sin(2 * Omega_r * time) * np.sin(omega_eg * time)

# Definir el rango de valores para la frecuencia de Rabi y el tiempo
Omega_r = 1.0  # Frecuencia de Rabi
omega_eg = 1.0  # Frecuencia de transición
M = 1.0  # Constante M
time_range = np.linspace(0, 10, 1000)  # Cambia los límites y la cantidad de puntos según sea necesario

# Calcular el valor esperado del operador dipolar para cada valor de tiempo
dipolar_values = dipolar_value(M, Omega_r, omega_eg, time_range)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(time_range, dipolar_values)
plt.xlabel('Tiempo')
plt.ylabel('Valor Esperado del Operador Dipolar')
plt.title('Valor Esperado del Operador Dipolar en Oscilaciones de Rabi')
plt.grid(True)
plt.show()
```
