```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
c = 3e8  # Velocidad de la luz en el vacío (m/s)
lambda_0 = 8e-6  # Longitud de onda central (m)
T0 = 1e-15  # Duración del pulso (s)
omega_0 = 2 * np.pi * c / lambda_0  # Frecuencia angular central (rad/s)

# Parámetros de simulación
z_max = 1e-3  # Distancia de propagación máxima (m)
Nz = 1000  # Número de pasos en z
z = np.linspace(0, z_max, Nz)  # Eje de propagación

# Tiempo y frecuencia
Nt = 1024  # Número de puntos en el tiempo
t_max = 5 * T0
t = np.linspace(-t_max, t_max, Nt)  # Eje de tiempo
dt = t[1] - t[0]
dw = 2 * np.pi / (Nt * dt)  # Resolución en frecuencia
omega = np.fft.fftfreq(Nt, d=dt) * 2 * np.pi  # Eje de frecuencia

# Parámetros de la fibra (ejemplo)
beta2 = -20e-27  # Dispersión de grupo (s^2/m)
gamma = 1e-3  # Coeficiente de no linealidad (W^-1 m^-1)

# Pulso inicial (Gaussiano)
A0 = np.exp(-t**2 / (2 * T0**2)) * np.exp(1j * omega_0 * t)

# Método de Split-Step Fourier
def NLS_propagate(A0, z, beta2, gamma, dt, dz):
    A = A0.copy()
    for i in range(len(z)):
        A_w = np.fft.fft(A) * np.exp(-1j * beta2 * omega**2 * dz / 2)
        A = np.fft.ifft(A_w)
        A *= np.exp(1j * gamma * np.abs(A)**2 * dz)
    return A

dz = z[1] - z[0]  # Paso en z
A_z = NLS_propagate(A0, z, beta2, gamma, dt, dz)

# Visualización del resultado
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(t, np.abs(A0)**2, label='z=0')
plt.plot(t, np.abs(A_z)**2, label=f'z={z_max:.2e} m')
plt.xlabel('Tiempo (s)')
plt.ylabel('Intensidad (a.u.)')
plt.legend()

plt.subplot(122)
A_w = np.fft.fftshift(np.fft.fft(A_z))
omega_shifted = np.fft.fftshift(omega)
plt.plot(omega_shifted, np.abs(A_w)**2)
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Espectro (a.u.)')

plt.tight_layout()
plt.show()
```

```python
def simulate_pulse_for_material(beta2, gamma, z, A0, omega, dt, dz):
    return NLS_propagate(A0, z, beta2, gamma, dt, dz)

# Parámetros para cada material
materials = {
    "Silica": {"beta2": 36e-27, "gamma": 1.2e-3},
    "Sapphire": {"beta2": 20e-27, "gamma": 0.9e-3},
    "Xenon": {"beta2": -10e-27, "gamma": 0.5e-3},
}

# Simulación para cada material
results = {}
for material, params in materials.items():
    beta2 = params["beta2"]
    gamma = params["gamma"]
    A_z = simulate_pulse_for_material(beta2, gamma, z, A0, omega, dt, dz)
    results[material] = A_z
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def jaynes_cummings(t, y, omega_a, omega_c, g, gamma_a, gamma_c):
    i=0
    # Estados del sistema: P_ee, P_eg, P_ge, P_gg
    P_ee, P_eg, P_ge, P_gg = y
    
    # Derivadas
    dP_ee_dt = -gamma_a * P_ee - 2 * g * np.imag(P_eg)
    dP_eg_dt = - (gamma_a + gamma_c) / 2 * P_eg + g * (P_gg - P_ee) - i * (omega_a - omega_c) * P_eg
    dP_ge_dt = np.conj(dP_eg_dt)
    dP_gg_dt = gamma_a * P_ee + gamma_c * (P_eg + P_ge) + 2 * g * np.imag(P_eg)
    
    return [dP_ee_dt, dP_eg_dt, dP_ge_dt, dP_gg_dt]

# Parámetros
omega_a = 1.0  # Frecuencia del átomo
omega_c = 1.0  # Frecuencia del campo
gamma_a = 0.1  # Tasa de decaimiento del átomo
gamma_c = 0.05  # Tasa de decaimiento del campo
t_max = 20
t_eval = np.linspace(0, t_max, 1000)

# Condiciones iniciales (átomo excitado, campo vacío)
y0 = [1, 0, 0, 0]

# Diferentes valores de la constante de acoplamiento
g_values = [0.1, 0.5, 1.0]

# Simulación y visualización
for g in g_values:
    sol = solve_ivp(jaynes_cummings, (0, t_max), y0, args=(omega_a, omega_c, g, gamma_a, gamma_c), t_eval=t_eval)
    plt.plot(t_eval, sol.y[0], label=f'g = {g}')

plt.xlabel('Tiempo')
plt.ylabel('Probabilidad de encontrar el átomo excitado')
plt.legend()
plt.title('Evolución de la población del átomo excitado para diferentes acoplamientos')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

def inversion_population(bar_n, gamma, t_values):
    """
    Calcula la inversión de población W(t) para un estado coherente del campo.

    Parameters:
        bar_n (float): Número promedio de fotones en el estado coherente.
        gamma (float): Acoplamiento (constante de Rabi).
        t_values (ndarray): Valores de tiempo para evaluar W(t).

    Returns:
        W_t (ndarray): Inversión de población en función del tiempo.
    """
    W_t = []
    max_terms = 200  # Número razonable de términos para la serie
    precomputed_factors = np.exp(-bar_n) * (bar_n**np.arange(max_terms)) / factorial(np.arange(max_terms))

    for t in t_values:
        sum_terms = sum(
            precomputed_factors[n] * np.cos(2 * gamma * t * np.sqrt(n + 1))
            for n in range(max_terms)
        )
        W_t.append(sum_terms)

    return np.array(W_t)

# Parámetros
bar_n = 5  # Número promedio de fotones
gamma = 1.0  # Constante de acoplamiento
t_values = np.linspace(0, 50, 1000)  # Tiempo

# Calcular W(t)
W_t = inversion_population(bar_n, gamma, t_values)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(t_values, W_t, label=f"$W(t)$ con $\\bar{{n}}={bar_n}$ y $\\gamma={gamma}$",)
plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.title("Evolución temporal de la inversión de población $W(t)$", fontsize=14)
plt.xlabel("Tiempo $t$", fontsize=12)
plt.ylabel("Inversión de población $W(t)$", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del pulso
A = 1.0          # Amplitud
tau = 2.0        # Duración del pulso (ancho)
t_rise = 0.1     # Tiempo de subida
t_fall = 0.1     # Tiempo de bajada
t0 = 5.0         # Centro del pulso

# Función para generar el pulso
def square_pulse(t, A, tau, t_rise, t_fall, t0):
    pulse = np.zeros_like(t)
    # Tiempos de subida/bajada
    rise = (t >= t0 - tau/2) & (t < t0 - tau/2 + t_rise)
    fall = (t >= t0 + tau/2 - t_fall) & (t < t0 + tau/2)
    flat = (t >= t0 - tau/2 + t_rise) & (t < t0 + tau/2 - t_fall)
    # Aplica subida exponencial
    pulse[rise] = A * (1 - np.exp(-(t[rise] - (t0 - tau/2)) / t_rise))
    # Parte plana
    pulse[flat] = A
    # Aplica bajada exponencial
    pulse[fall] = A * np.exp(-(t[fall] - (t0 + tau/2 - t_fall)) / t_fall)
    return pulse

# Rango de tiempo
t = np.linspace(0, 10, 1000)
pulso = square_pulse(t, A, tau, t_rise, t_fall, t0)

# Gráfico
plt.figure(figsize=(10, 5))
plt.plot(t, pulso, linewidth=2)
plt.xlabel("Tiempo ($t$)", fontsize=12)
plt.ylabel("Amplitud ($A$)", fontsize=12)
plt.title("Pulso cuadrado", fontsize=14)
plt.grid(True )
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Parámetros del sistema
omega_c = 1.0  # Frecuencia de la cavidad
omega_a = 1.0  # Frecuencia del átomo
g0 = 1.0       # Amplitud del pulso
t0 = 1.0       # Duración del pulso
T_R = 10.0     # Período de repetición del pulso
n_pulses = 5   # Número de pulsos

# Función que define el tren de pulsos g(t)
def g(t):
    # Calcular el índice del pulso más cercano
    n = int(np.round(t / T_R))
    return g0 * np.exp(-((t - n * T_R) / t0)**2)

# Ecuación de von Neumann
def von_neumann(t, rho, omega_c, omega_a, g0, t0, T_R):
    rho = rho.reshape(3, 3)
    g_t = g(t)
    
    # Matriz de la ecuación de von Neumann
    d_rho = -1j * np.array([
        [0, -omega_c * rho[0, 1] - g_t * rho[0, 2], -g_t * rho[0, 1] - omega_a * rho[0, 2]],
        [omega_c * rho[1, 0] + g_t * rho[2, 0], g_t * (rho[2, 1] - rho[1, 2]), g_t * (rho[2, 2] - rho[1, 1]) + rho[1, 2] * (omega_c - omega_a)],
        [g_t * rho[0, 1] + omega_a * rho[2, 0], g_t * (rho[1, 1] - rho[2, 2]) + rho[2, 1] * (omega_a - omega_c), g_t * (rho[1, 2] - rho[2, 1])]
    ])
    
    return d_rho.flatten()

# Condiciones iniciales
rho0 = np.zeros((3, 3), dtype=complex)
rho0[1, 1] = 1.0  # Estado inicial

# Tiempo de integración
t_span = (0, n_pulses * T_R)
t_eval = np.linspace(0, n_pulses * T_R, 1000)

# Resolver la ecuación diferencial
sol = solve_ivp(von_neumann, t_span, rho0.flatten(), t_eval=t_eval, args=(omega_c, omega_a, g0, t0, T_R), method='RK45')

# Extraer la solución
rho_t = sol.y.reshape(3, 3, -1)

# Graficar los resultados
plt.figure()
plt.plot(sol.t, np.abs(rho_t[1, 2, :]), label='rho_23')
plt.plot(sol.t, np.abs(rho_t[1, 1, :]), label='rho_22')
plt.plot(sol.t, 0*rho_t[0, 0, :]+ 0*rho_t[1, 1, :]+1*rho_t[2, 2, :] , label='rho_22')
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

# Parámetros del sistema
omega_c = 1.0  # Frecuencia de la cavidad
omega_a = 1.0  # Frecuencia del átomo
g0 = 1.0       # Amplitud del pulso
t0 = 1.0       # Duración del pulso
T_R = 10.0     # Período de repetición del pulso
n_pulses = 5   # Número de pulsos

# Función que define el tren de pulsos g(t)
def g(t):
    # Calcular el índice del pulso más cercano
    n = int(np.round(t / T_R))
    return g0 * np.exp(-((t - n * T_R) / t0)**2)

# Ecuación de von Neumann
def von_neumann(t, rho, omega_c, omega_a, g0, t0, T_R):
    rho = rho.reshape(3, 3)
    g_t = g(t)
    
    # Matriz de la ecuación de von Neumann
    d_rho = -1j * np.array([
        [0, -omega_c * rho[0, 1] - g_t * rho[0, 2], -g_t * rho[0, 1] - omega_a * rho[0, 2]],
        [omega_c * rho[1, 0] + g_t * rho[2, 0], g_t * (rho[2, 1] - rho[1, 2]), g_t * (rho[2, 2] - rho[1, 1]) + rho[1, 2] * (omega_c - omega_a)],
        [g_t * rho[0, 1] + omega_a * rho[2, 0], g_t * (rho[1, 1] - rho[2, 2]) + rho[2, 1] * (omega_a - omega_c), g_t * (rho[1, 2] - rho[2, 1])]
    ])
    
    return d_rho.flatten()

# Función para resolver y graficar
def plot_interactive(omega_c, omega_a, g0, t0, T_R, n_pulses):
    # Condiciones iniciales
    rho0 = np.zeros((3, 3), dtype=complex)
    rho0[1, 1] = 1.0  # Estado inicial

    # Tiempo de integración
    t_span = (0, n_pulses * T_R)
    t_eval = np.linspace(0, n_pulses * T_R, 1000)

    # Resolver la ecuación diferencial
    sol = solve_ivp(von_neumann, t_span, rho0.flatten(), t_eval=t_eval, args=(omega_c, omega_a, g0, t0, T_R), method='RK45')

    # Extraer la solución
    rho_t = sol.y.reshape(3, 3, -1)

    # Graficar los resultados
    plt.figure(figsize=(10, 6))
    plt.plot(sol.t, np.abs(rho_t[1, 2, :]), label='rho_23')
    plt.plot(sol.t, np.abs(rho_t[1, 1, :]), label='rho_22')
    plt.plot(sol.t, 0*rho_t[0, 0, :]+ 0*rho_t[1, 1, :]+1*rho_t[2, 2, :] , label='rho_Diagonal')
    plt.xlabel('Tiempo')
    plt.ylabel('Población')
    plt.legend(
        
    )
    plt.show()

# Crear controles interactivos
interact(plot_interactive,
         omega_c=FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='omega_c:'),
         omega_a=FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='omega_a:'),
         g0=FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='g0:'),
         t0=FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='t0:'),
         T_R=FloatSlider(value=10.0, min=5.0, max=20.0, step=1.0, description='T_R:'),
         n_pulses=IntSlider(value=5, min=1, max=10, step=1, description='n_pulses:'))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la cavidad
L = 1.0  # Longitud de la cavidad (de z=0 a z=L)
k = 4 * np.pi / L  # Número de onda para una onda estacionaria en la cavidad
A = 1.0  # Amplitud de la onda

# Función de la onda sinusoidal en la cavidad
def onda_sinusoidal(z, t=0):
    return A * np.sin(k * z) * np.cos(omega * t)

# Frecuencia angular de la onda (relacionada con el número de onda k)
omega = k  # Para simplificar, usamos omega = k

# Crear un array de valores de z desde 0 hasta L
z_values = np.linspace(0, L, 500)

# Calcular la onda en el instante t = 0
t = 0
onda_values = onda_sinusoidal(z_values, t)

# Graficar la onda en la cavidad
plt.figure(figsize=(10, 6))
plt.plot(z_values, onda_values, color='black')
plt.title('Onda en una cavidad entre Z=0 y Z=L')
plt.xlabel('Posición (Z)')
plt.ylabel('Amplitud')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Línea horizontal en y=0
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Línea vertical en z=0
plt.axvline(L, color='black', linewidth=0.5, linestyle='--')  # Línea vertical en z=L
plt.legend()
plt.grid(False)
plt.show()
```
