```python
import numpy as np
from scipy.optimize import curve_fit

# Función gaussiana
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) / stddev) ** 2)

# Creamos una función para el cálculo del FWHM
def calculate_fwhm(x, y):
    half_max = max(y) / 2.0
    idx_left = np.where(y > half_max)[0][0]
    idx_right = np.where(y > half_max)[0][-1]
    fwhm = x[idx_right] - x[idx_left]
    return fwhm

# Creamos listas para almacenar los anchos de cada gaussiana
widths = []

# Iteramos sobre cada pico para ajustar una gaussiana y calcular su ancho
for i in range(len(peaks_8)):
    # Definimos un rango para el ajuste de la gaussiana alrededor de cada pico
    peak_range = 20  # Rango alrededor del pico
    peak_position = peaks_8[i]  # Posición del pico
    
    peak_start = max(0, peak_position - peak_range)
    peak_end = min(len(x_values) - 1, peak_position + peak_range)
    
    x_fit = x_values[peak_start:peak_end]
    y_fit = y_values[peak_start:peak_end]
    
    # Estimamos parámetros iniciales para la gaussiana
    amplitude_guess = max(y_fit) - min(y_fit)
    mean_guess = x_values[peaks_8[i]]
    stddev_guess = np.std(y_fit)
    
    # Ajustamos una gaussiana a un pico específico
    popt, pcov = curve_fit(gaussian, x_fit, y_fit, p0=[amplitude_guess, mean_guess, stddev_guess])
    
    # Calculamos el ancho a media altura (FWHM) de la gaussiana ajustada
    x_fit = np.linspace(x_values[peak_start], x_values[peak_end - 1], 100)
    y_fit = gaussian(x_fit, *popt)
    
    fwhm = calculate_fwhm(x_fit, y_fit)
    widths.append(fwhm)

# Creamos una tabla con los anchos de cada gaussiana
import pandas as pd

data = {'Pico': [f'Pico {i+1}' for i in range(len(peaks_8))],
        'Ancho FWHM': widths}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks

# Cargar el archivo CSV
file_path = 'C:\\Users\\madro\\OneDrive\\Documents\\.py\\Python-\\TecnicasDeCaractenizacion\\Data\\PowDLL_XY_XRD_Sunday, December 03, 2023.csv'
data = pd.read_csv(file_path)

# Obtener los valores de las columnas 'Eje X' y 'Eje Y'
x_values = data['2Theta'].tolist()
y_values = data['cuenta'].tolist()

# Función de suavizado por promedio avanzado
def advanced_averaging(data, window_size):
    smoothed_data = []
    half_window = window_size // 2
    
    for i in range(len(data)):
        if i < half_window or i >= len(data) - half_window:
            smoothed_data.append(data[i])
        else:
            window = data[i - half_window: i + half_window + 1]
            smoothed_value = sum(window) / window_size
            smoothed_data.append(smoothed_value)
    
    return smoothed_data

# Aplicar suavizado a los datos 'y_values'
window_size = 5  # Modifica el tamaño de la ventana según lo deseado
y_smoothed = advanced_averaging(y_values, window_size)

# Graficar los datos suavizados
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_smoothed, 'b-')
plt.xlabel('2$\\theta$')
plt.ylabel('Intensidad (Cuentas)')
#plt.title('Gráfico de datos suavizados')
plt.grid(True)

# Encontrar todos los picos en los datos suavizados
peaks, _ = find_peaks(y_smoothed, distance=20, prominence=0.5)

# Ordenar los picos por altura
sorted_peaks = sorted(peaks, key=lambda x: y_smoothed[x], reverse=True)

# Seleccionar los 15 picos más altos
peaks_15 = sorted_peaks[:15]

# Marcar los picos en el gráfico de datos suavizados
plt.plot([x_values[i] for i in peaks_15], [y_smoothed[i] for i in peaks_15], 'ro', label='Picos')

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Cargar los datos desde el archivo CSV
file_path = 'C:\\Users\\madro\\OneDrive\\Documents\\.py\\Python-\\TecnicasDeCaractenizacion\\Data\\W-H.csv'  # Reemplaza 'tu_archivo.csv' con la ruta de tu archivo CSV
data = pd.read_csv(file_path)

# Extraer las columnas necesarias para el ajuste (suponiendo 'x' y 'y')
x = data['4 SEN (TETA)']
y = data['BETA COS (TETA)']

# Realizar el ajuste lineal
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Calcular la recta de ajuste
linea_ajuste = slope * x + intercept

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Datos')
plt.plot(x, linea_ajuste, 'r', label='Ajuste Lineal')

# Limitar el rango en el eje Y
plt.ylim(0.006, 0.012)


# Mostrar la ecuación del ajuste en la gráfica
ecuacion_ajuste = f'Ecuación del ajuste: y = {slope:.5f}x + {intercept:.5f}\nR-cuadrado: {r_value**2:.2f}'
plt.text(0.5, 0.5, ecuacion_ajuste, fontsize=12, ha='left', va='center', transform=plt.gca().transAxes)

# Configuraciones adicionales
plt.xlabel('$4sen\\theta$')
plt.ylabel('$\\beta cos \\theta$')
plt.title('Williamson - Hall model')
plt.legend()
plt.grid(True)
plt.show()
```
